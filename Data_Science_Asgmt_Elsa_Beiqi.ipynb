# 8. f. Split the dataset into training and testing, with the split date being January 1, 2022.


```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# Extract year and week number as features
df['Year'] = df['WEEK_START'].dt.isocalendar().year
df['Week'] = df['WEEK_START'].dt.isocalendar().week

# Split into train and test sets
train = df[df['WEEK_START'] < '2022-01-01']
test = df[df['WEEK_START'] >= '2022-01-01']

# Fit ARIMA model on train set
model = ARIMA(train['average_rent'], order=(1, 0, 0))
model_fit = model.fit()

# Make predictions on train and test set
train_pred = model_fit.predict(start=1, end=len(train)-1)
test_pred = model_fit.predict(start=len(train), end=len(df)-1)

# Calculate and print RMSE for train and test set
train_rmse = np.sqrt(mean_squared_error(train['average_rent'][1:], train_pred))
test_rmse = np.sqrt(mean_squared_error(test['average_rent'], test_pred))
print("Train RMSE: {:.2f}".format(train_rmse))
print("Test RMSE: {:.2f}".format(test_rmse))

# Plot the graph
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(train['WEEK_START'], train['average_rent'], label='Train')
ax.plot(test['WEEK_START'], test['average_rent'], label='Test')
ax.set_xlabel('Date')
ax.set_ylabel('Average Rent')
ax.legend()
plt.show()
```

1.  I used ARIMA, which is a time series forecasting model, due to the nature of the problem and the available data. Time series models are specifically designed to work with data that has a temporal component and can capture seasonality, trends, and other patterns that may exist in the data.
2. The test RMSE is higher than the train RMSE, which suggests that the model may be overfitting to the training data.

```python
import statsmodels.api as sm

# Fit an exponential smoothing model on train set
model = sm.tsa.ExponentialSmoothing(train, seasonal_periods=52, seasonal='add', initialization_method='estimated')
fit = model.fit(smoothing_level=0.3, smoothing_trend=0.05, smoothing_seasonal=0.2, optimized=False)

# Forecast one year into the future (i.e. for 2023)
future_pred = fit.forecast(52)

# Plot the forecasted values for 2023
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(pd.date_range(start='2023-01-01', periods=len(future_pred), freq='W'), future_pred, label='Forecast')
ax.set_xlabel('Date')
ax.set_ylabel('Average Rent')
ax.set_title('Average Rent Forecast for 2023 (Exponential Smoothing)')
ax.legend()
plt.show()
```

3.
*   I used Exponential Smoothing because of its ability to handle seasonal patterns in data. It is based on the idea that more recent observations should be given more weight in forecasting future observations. I used it since it is a simple and efficient method for forecasting.
* The forecast is consistent with historical trends.

```python
from sklearn.ensemble import RandomForestRegressor

# Define the independent and dependent variables for training and testing
X_train = train[['total_population', 'total_household', 'physical_visits', 'virtual_visits']]
y_train = train['average_rent']
X_test = test[['total_population', 'total_household', 'physical_visits', 'virtual_visits']]
y_test = test['average_rent']

# Create a Random Forest Regressor model
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions on train and test set
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calculate and print RMSE for train and test set
train_rmse = np.sqrt(mean_squared_error(y_train, model.predict(X_train)))
test_rmse = np.sqrt(mean_squared_error(y_test, model.predict(X_test)))

print("Train RMSE: {:.2f}".format(train_rmse))
print("Test RMSE: {:.2f}".format(test_rmse))

# Forecast for 2023
X_2023 = np.array([df.iloc[-1][['total_population', 'total_household', 'physical_visits', 'virtual_visits']]])
average_rent_2023 = model.predict(X_2023)[0]

print(f"The forecasted average rent for 2023 is {average_rent_2023:.2f} dollars.")
```
